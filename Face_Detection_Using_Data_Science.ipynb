```
import numpy as np
import matplotlib.pyplot as plt
```

```
img = plt.imread('/content/IMG-20171021-WA0017.jpg')
plt.imshow(img)
plt.axis('off')
```

```
img.shape
```

```
import cv2 
```

```
model = cv2.CascadeClassifier('/content/haarcascade_frontalface_alt.xml')
faces = model.detectMultiScale(img)
```

```
faces.shape
```

```
for face in faces:
  x,y,w,h = face

  img_box = cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0), 2)
```

```
plt.imshow(img_box)
plt.title('Faces Detected!!')
```
